```python
import os
import numpy as np
import cv2
from PIL import Image

original_path = './tsrd-train/'
resized_path = './aa/'

file_list = os.listdir(original_path)
img_list = []

for item in file_list :
    if item.find('.png') is not -1 :
        img_list.append(item)

total_image = len(img_list)
index = 0

for name in img_list :

    img = Image.open('%s%s'%(original_path, name))
    img_array = np.array(img)
    img_resize = cv2.resize(img_array, (100,100), interpolation = cv2.INTER_AREA)
    img = Image.fromarray(img_resize)
    img.save('%s%s.PNG'%(resized_path, name))

    print(name + '   ' + str(index) + '/' + str(total_image))
    index = index + 1
```

```python
import glob
file_list_train = []
for filename in glob.glob('./aa/*.png'):
    file_list_train.append(filename)
    print(filename)
```

```python
images_train = []
for image in file_list_train:
    images = cv2.imread(image)
    images_train.append(images)
```

```python
X_train = np.stack(images_train)
print(X_train.shape)
```

```python
file_list_test=[]
for filename in glob.glob('./tsrd-test/*.png'):
    images = cv2.imread(filename)
    image_resized = cv2.resize(images, dsize=(100, 100))
    file_list_test.append(image_resized)
```

```python
X_test = np.stack(file_list_test)
X_test
```

```python
import pandas as pd

y_train = pd.read_csv("./tsrd-train/TsignRecgTrain4170Annotation.txt", sep=';')
y_test = pd.read_csv("./tsrd-test/TsignRecgTest1994Annotation.txt", sep=';')
```

```python
y_train = y_train.iloc[:, 7].sort_values()
```

```python
y_test = y_test.iloc[:, 7].sort_values()
```

```python
y_train.reset_index(drop=True, inplace =True )
```

```python
y_test.reset_index(drop=True, inplace =True )
```

```python
X_train_centered = X_train/255
X_test_centered = X_test/255
```

```python
X_train_centered= X_train_centered.reshape((-1,100,100,3))
X_test_centered= X_train_centered.reshape((-1,100,100,3))
```

```python
from tensorflow.keras.utils import to_categorical

y_train_onehot = to_categorical(y_train)
y_test_onehot = to_categorical(y_test)
```

## train CNN

```python
from tensorflow.keras import layers, models
model = models.Sequential()
model.add(layers.Conv2D(32, (5,5), padding='valid', activation='relu', input_shape = (100,100,3)))
model.add(layers.MaxPool2D((2,2)))
```

```python
model.add(layers.Conv2D(64, (5,5), padding='valid', activation='relu', input_shape = (100,100,3)))
model.add(layers.MaxPool2D((2,2)))
```

```python
model.add(layers.Conv2D(64, (5,5), padding='valid', activation='relu', input_shape = (100,100,3)))
model.add(layers.MaxPool2D((2,2)))

model.add(layers.Flatten())
```

```python
model.add(layers.Dense(1024, activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(58, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer='adam', metrics=['acc'])
```

```python
import time
from tensorflow.keras.callbacks import ModelCheckpoint, TensorBoard
```

```python
from datetime import datetime
LOG_DIRECTORY_ROOT=""
now = datetime.utcnow().strftime("%Y%m%d%H%M%S")
log_dir="{}/run-{}/".format(LOG_DIRECTORY_ROOT,now)
callback_list = [ModelCheckpoint(filepath = 'cnn_checkpoint.h5',
                                 monitor='val_loss',
                                 save_best_only = True,
                                 TensorBoard = TensorBoard(log_dir=log_dir, write_graph=True, write_images=True))]
```

```python
history = model.fit(X_train_centered, y_train_onehot,
                   batch_size =50, epochs = 10,
                   validation_data=(X_train_centered, y_train_onehot),
                   callbacks=callback_list)
```

```python
import matplotlib.pyplot as plt
epochs = np.arange(1,11)
plt.plot(epochs, history.history['loss'])
plt.plot(epochs, history.history['val_loss'])
plt.xlabel('epochs')
plt.ylabel('loss')
plt.show()
```

```python
plt.plot(epochs, history.history['acc'])
plt.plot(epochs, history.history['val_acc'])
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.show()
```
